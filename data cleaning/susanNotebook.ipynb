Import Dependencies and Data

```python
# Import dependencies
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
# Import and read data
crash_df =pd.read_csv("Airplane_Crashes_And_Fatalities_Since_1908 (2).csv")
crash_df
```

```python
# Select useful columns
columns_df = crash_df.loc[:, ["Date", "Location", "Operator", "Aboard", "Fatalities", "Summary"]]
columns_df.tail()
```

```python
# Divide date into 3 parts
columns_df[['month','day','year']] = columns_df.Date.str.split('/', expand=True)
columns_df
```

```python
# Divide location into 4 parts
columns_df[['city','state','country','extra']] = columns_df.Location.str.split(',', expand=True)
columns_df
```

```python
# Drop unneeded columns
df2 = columns_df.drop(['Date', 'day', 'month'], axis=1)
```
